```python
import pandas as pd
from lifelines import CoxPHFitter 
import numpy as np

data=pd.read_csv('/home/alanah/Downloads/tmb_mskcc_2018_clinical_data.tsv',sep='\t', index_col=0)
print(data.head())

col_names = list(data.columns.values)
print(col_names)
data = data.astype({'Overall Survival (Months)':'float'})
print(data['Overall Survival (Months)'])

cph = CoxPHFitter()
cph.fit(data,'Overall Survival (Months)', event_col='Overall Survival Status')

cph.print_summary()


```
